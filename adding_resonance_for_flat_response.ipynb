## Using resonance to flatten frequency response

Resonance can be used to improve the flatness of the frequency response of a low pass filter.


```python
%matplotlib
```

```python
f=logspace(0,4,1000)
#f=linspace(0,1000,1000)
s=1j*2*pi*f

fc=100
wc=100*2*pi
B=50*2*pi
K=wc**2
n=4
H1=10*(wc/(s+wc))**(n-2) * K/(s**2 + B*s + K)
H2=10*(wc/(s+wc))**n 
ylim((-40,30))
semilogx(f,20*log10(abs(H1)),f,20*log10(abs(H2)))
grid()
legend(('resonance','no res.'));
```
