```python
%matplotlib inline
%pylab inline
ω=logspace(-3,3)
L,C=1,1
Z=ω*L-1/(ω*C)
semilogx(ω,Z);
grid()
```

```python
ω=linspace(0.1,10,100)
L,C=1,1
Z=ω*L-1/(ω*C)
plot(ω,Z);
grid()
```

```python
R,L,C=5,1,0.25
Vs=10*exp(1j*0)
ω=2
f=ω/(2*pi)
T=1/f
Zl=1j*ω*L
Zc=1/(1j*ω*C)
Zr=R
Ztotal=Zl+Zc+Zr

Vr=Vs * ( Zr/Ztotal)
Vc=Vs * ( Zc/Ztotal)
Vl=Vs * ( Zl/Ztotal)
print(Vl,Vc,Vr)
t=linspace(0,5*T,1000)
vs=abs(Vs)*cos(ω*t+angle(Vs))
vc=abs(Vc)*cos(ω*t+angle(Vc))
vr=abs(Vr)*cos(ω*t+angle(Vr))
vl=abs(Vl)*cos(ω*t+angle(Vl))

vtotal=vc+vr+vl
plot(t,vs,t,vc,t,vr,t,vl);
legend(('Vs','C','R','L'))
```
