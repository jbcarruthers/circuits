```python
%pylab inline
f=logspace(0,6,1000)
s=1j*2*pi*f
```

```python
H=10/((1+s/(2*pi*100)))
#plot(f,abs(H));
#H=1*(f<100)
#plot(f,H)
a=100*2*pi
b=100*2*pi
K=10
n=4
H2=K * (a/(s+a))**n
```

```python
figure(figsize=(10,6))
semilogx(f,20*log10(abs(H)),f,20*log10(abs(H2)))
xlabel('Frequency (Hz)')
ylim((-100,30))
grid()
```
